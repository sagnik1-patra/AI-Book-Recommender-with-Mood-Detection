```python
# -*- coding: utf-8 -*-
# Interactive: user types an emotion  get book recommendations
# Artifacts required (already created earlier):
#   C:\Users\sagni\Downloads\AI Book Recommender with Mood Detection\bx_artifacts.pkl
#   C:\Users\sagni\Downloads\AI Book Recommender with Mood Detection\bx_interactions.h5

import os, json, pickle
from pathlib import Path

import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix

# =========================
# CONFIG  edit if needed
# =========================
BASE_DIR = r"C:\Users\sagni\Downloads\AI Book Recommender with Mood Detection"
PKL  = str(Path(BASE_DIR) / "bx_artifacts.pkl")
H5   = str(Path(BASE_DIR) / "bx_interactions.h5")

TOP_N    = 20     # number of recommendations to show
SIM_TOPK = 2000   # neighbors considered in user-based CF
BLEND_W  = 0.85   # CF vs popularity blend (0..1)

# Optional: target a specific Book-Crossing user id; else auto-pick an active user
USER_ID_OVERRIDE = None   # e.g., "276729" or leave as None

# =========================
# Load artifacts
# =========================
if not Path(PKL).exists() or not Path(H5).exists():
    raise SystemExit("Artifacts not found. Build them first (bx_artifacts.pkl, bx_interactions.h5).")

with open(PKL, "rb") as f:
    ART = pickle.load(f)

with h5py.File(H5, "r") as hf:
    shp = tuple(hf["ratings_binary"].attrs["shape"])
    Rb = csr_matrix((hf["ratings_binary"]["data"][:],
                     hf["ratings_binary"]["indices"][:],
                     hf["ratings_binary"]["indptr"][:]), shape=shp)

users       = ART["users"]
items_isbn  = ART["items_isbn"]
items_title = ART["items_title"]
items_author= ART["items_author"]
items_year  = ART["items_year"]
items_cat   = ART["items_category"]
items_lang  = ART["items_language"]

n_users, n_items = Rb.shape
print(f"[INFO] Matrix loaded: users={n_users:,}, items={n_items:,}, positives={Rb.nnz:,}")

# =========================
# Helpers
# =========================
def trunc(s, n=40):
    s = s or ""
    return s if len(s) <= n else s[:n-1] + ""

def row_l2norms_csr(M: csr_matrix):
    sq = np.array(M.power(2).sum(axis=1)).ravel()
    norms = np.sqrt(np.maximum(sq, 1e-12))
    return norms

def normalize_rows_csr(M: csr_matrix):
    norms = row_l2norms_csr(M)
    inv = 1.0 / norms
    Mn = M.multiply(inv.reshape(-1,1))
    return Mn.tocsr(), norms

# Emotion  genre/category keywords (customize freely)
EMOTION_TO_GENRES = {
    "happy":      ["humor", "comedy", "light", "feel-good", "uplifting"],
    "sad":        ["drama", "literary", "tragedy", "poignant"],
    "lonely":     ["literary", "drama", "introspective"],
    "anxious":    ["self-help", "mindfulness", "inspirational", "psychology"],
    "motivated":  ["self-help", "business", "productivity", "inspirational"],
    "romantic":   ["romance", "love", "relationship", "chick lit"],
    "bored":      ["adventure", "fantasy", "mystery", "thriller"],
    "curious":    ["non-fiction", "science", "history", "biography"],
    "adventurous":["adventure", "travel", "fantasy", "action"],
    "scared":     ["horror", "thriller", "suspense"],
    "calm":       ["poetry", "mindfulness", "philosophy"],
    "inspired":   ["biography", "memoir", "inspirational"],
    "nostalgic":  ["classic", "historical", "coming-of-age"],
    "mystery":    ["mystery", "detective", "crime", "thriller"],
    "fantasy":    ["fantasy", "magic", "dragon"],
    "scifi":      ["science fiction", "sci-fi", "sf"],
}

def derive_allowed_categories(emotion_text: str):
    t = (emotion_text or "").strip().lower()
    allow = set()
    # map known emotion keywords
    for emo, cats in EMOTION_TO_GENRES.items():
        if emo in t:
            allow.update(cats)
    # also include tokens from free text
    for tok in t.replace(",", " ").split():
        if len(tok) >= 3:
            allow.add(tok)
    return allow if allow else None

def build_candidate_mask(allowed_cats):
    mask = np.ones(n_items, dtype=bool)
    if allowed_cats:
        cat_series = pd.Series(items_cat, dtype=str).str.lower().fillna("")
        m = np.zeros(n_items, dtype=bool)
        for token in allowed_cats:
            m |= cat_series.str.contains(token, regex=False).values
        if m.sum() >= 50:   # require at least some breadth; else fallback
            mask &= m
        else:
            print("[WARN] Mood filter too strict; using all items.")
    return mask

# Popularity vector (for blending)
item_pop = np.asarray(Rb.getnnz(axis=0)).astype(np.float32)
if item_pop.max() > 0:
    item_pop /= item_pop.max()

# Normalize rows for cosine space (once)
Rb_norm, _ = normalize_rows_csr(Rb)

def recommend_by_emotion(emotion_text: str, user_id: str|None = None, top_n: int = 20,
                         sim_topk: int = 2000, blend_w: float = 0.85):
    # choose user
    user_nnz = np.asarray(Rb.getnnz(axis=1))
    if user_id is None:
        active = np.where(user_nnz >= 10)[0]
        if len(active) == 0:
            active = np.where(user_nnz >= 1)[0]
        if len(active) == 0:
            raise SystemExit("No users with interactions found.")
        u = int(active[0])
        uid = users[u]
    else:
        try:
            u = users.index(str(user_id))
        except ValueError:
            raise SystemExit(f"USER_ID '{user_id}' not found.")
        uid = users[u]

    print(f"[INFO] Recommending for user {uid} (row {u}) with {user_nnz[u]} interactions")

    allowed = derive_allowed_categories(emotion_text)
    cand_mask = build_candidate_mask(allowed)

    # target normalized row
    s, e = Rb_norm.indptr[u], Rb_norm.indptr[u+1]
    row_idx = Rb_norm.indices[s:e]
    row_dat = Rb_norm.data[s:e]
    u_vec = csr_matrix((row_dat, (np.zeros_like(row_idx), row_idx)), shape=(1, n_items))

    # cosine sim to all users
    sims = Rb_norm.dot(u_vec.T).toarray().ravel()
    sims[u] = 0.0

    # top-K neighbors
    if sim_topk < len(sims):
        top_idx = np.argpartition(-sims, sim_topk)[:sim_topk]
        sims_mask = np.zeros_like(sims, dtype=bool)
        sims_mask[top_idx] = True
        sims_filtered = np.where(sims_mask, sims, 0.0)
    else:
        sims_filtered = sims

    # user-based scores
    scores_cf = sims_filtered @ Rb
    scores_cf = np.array(scores_cf).ravel().astype(np.float32)

    # blend with popularity
    scores = blend_w * scores_cf + (1.0 - blend_w) * item_pop

    # mask out seen & apply candidate mask
    ss, ee = Rb.indptr[u], Rb.indptr[u+1]
    seen = set(Rb.indices[ss:ee].tolist())
    scores_masked = scores.copy()
    if seen:
        scores_masked[list(seen)] = -1e9
    scores_masked[~cand_mask] = -1e9

    # top-N safe
    available = int((scores_masked > -1e9).sum())
    k = min(top_n, max(1, available))
    top_idx = np.argpartition(-scores_masked, k-1)[:k]
    top_idx = top_idx[np.argsort(-scores_masked[top_idx])]

    # assemble dataframe
    rows = []
    for it in top_idx:
        yr = items_year[it]
        if isinstance(yr, (np.integer, int, np.int64, np.int32)): yr = int(yr)
        rows.append({
            "rank": len(rows)+1,
            "isbn": items_isbn[it],
            "title": items_title[it],
            "author": items_author[it],
            "year": yr,
            "category": items_cat[it],
            "language": items_lang[it],
            "score_cf": float(scores_cf[it]),
            "score_pop": float(item_pop[it]),
            "score_blend": float(scores[it]),
        })
    rec_df = pd.DataFrame(rows)

    # save CSV/JSON/PNG with emotion tag
    tag = (emotion_text.strip().lower().replace(" ", "_") or "all")
    csv_path  = os.path.join(BASE_DIR, f"recommendations_{tag}_{uid}.csv")
    json_path = os.path.join(BASE_DIR, f"recommendations_{tag}_{uid}.json")
    png_path  = os.path.join(BASE_DIR, f"Recommendations_{tag}_{uid}.png")

    rec_df.to_csv(csv_path, index=False, encoding="utf-8")
    with open(json_path, "w", encoding="utf-8") as f:
        json.dump({
            "user_id": uid,
            "emotion_text": emotion_text,
            "top_n": int(k),
            "blend_weight_cf": float(blend_w),
            "sim_topk_users": int(sim_topk),
            "recommendations": rec_df.to_dict(orient="records")
        }, f, ensure_ascii=False, indent=2)

    # bar chart
    plt.figure(figsize=(12, 6))
    labels = [trunc(t, 36) for t in rec_df["title"].tolist()]
    y = rec_df["score_blend"].values
    x = np.arange(len(labels))
    plt.bar(x, y)
    plt.xticks(x, labels, rotation=45, ha="right")
    plt.ylabel("Recommendation Score (blend)")
    title_extra = f" | Mood: {emotion_text}" if str(emotion_text).strip() else ""
    plt.title(f"Top-{k} Books for User {uid}{title_extra}")
    plt.tight_layout()
    plt.savefig(png_path, dpi=180, bbox_inches="tight")
    plt.show()

    print(f"[OK] Saved:\n  {csv_path}\n  {json_path}\n  {png_path}")
    print("\n=== Top Recommendations ===")
    for _, r in rec_df.head(10).iterrows():
        yr = f" ({r['year']})" if isinstance(r['year'], (int, np.integer)) and r['year']>0 else ""
        cat = f"  {r['category']}" if isinstance(r['category'], str) and r['category'] else ""
        print(f"{int(r['rank']):>2}. {r['title']}{yr}  by {r['author']}{cat}")

    return rec_df

# =========================
#  Interactive prompt
# =========================
try:
    emotion = input("Enter your emotion / mood (e.g., happy, sad, motivated mystery): ").strip()
except Exception:
    # Fallback for environments without stdin
    emotion = "happy"
    print(f"[WARN] No stdin; using default emotion: {emotion}")

# pick a specific user (optional)
uid_override = USER_ID_OVERRIDE

_ = recommend_by_emotion(
    emotion_text=emotion,
    user_id=uid_override,
    top_n=TOP_N,
    sim_topk=SIM_TOPK,
    blend_w=BLEND_W
)
```
