```python
# -*- coding: utf-8 -*-
# Book-Crossing â€” Prediction + Save Results + Chart (fixed CSR normalization)

import os, json, pickle
from pathlib import Path

import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix

# =========================
# CONFIG â€” edit if needed
# =========================
BASE_DIR   = r"C:\Users\sagni\Downloads\AI Book Recommender with Mood Detection"
USER_ID    = None    # put a specific BX "User-ID" string to target; else auto-pick an active user
MOOD_TEXT  = ""      # optional mood/genre text, e.g., "mystery", "motivated", "romance"
TOP_N      = 20      # number of recs
SIM_TOPK   = 2000    # most similar users to consider
BLEND_W    = 0.85    # blend weight for CF vs popularity (0..1)

PKL  = str(Path(BASE_DIR) / "bx_artifacts.pkl")
H5   = str(Path(BASE_DIR) / "bx_interactions.h5")

# =========================
# Load artifacts
# =========================
if not Path(PKL).exists() or not Path(H5).exists():
    raise SystemExit("Artifacts not found. Build them first (bx_artifacts.pkl, bx_interactions.h5).")

with open(PKL, "rb") as f:
    ART = pickle.load(f)

with h5py.File(H5, "r") as hf:
    shp = tuple(hf["ratings_binary"].attrs["shape"])
    Rb = csr_matrix((hf["ratings_binary"]["data"][:],
                     hf["ratings_binary"]["indices"][:],
                     hf["ratings_binary"]["indptr"][:]), shape=shp)

users       = ART["users"]
items_isbn  = ART["items_isbn"]
items_title = ART["items_title"]
items_author= ART["items_author"]
items_year  = ART["items_year"]
items_cat   = ART["items_category"]
items_lang  = ART["items_language"]

n_users, n_items = Rb.shape
print(f"[INFO] Loaded implicit matrix: users={n_users:,}, items={n_items:,}, positives={Rb.nnz:,}")

# =========================
# Pick target user
# =========================
user_nnz = np.asarray(Rb.getnnz(axis=1))
active_users = np.where(user_nnz >= 10)[0]  # at least 10 interactions
if USER_ID is None:
    if len(active_users) == 0:
        active_users = np.where(user_nnz >= 1)[0]
    if len(active_users) == 0:
        raise SystemExit("No users with interactions found.")
    target_u = int(active_users[0])
    USER_ID = users[target_u]
else:
    try:
        target_u = users.index(str(USER_ID))
    except ValueError:
        raise SystemExit(f"USER_ID '{USER_ID}' not found. Try USER_ID=None to auto-pick.")

print(f"[INFO] Target user: {USER_ID} (row {target_u}) with {user_nnz[target_u]} interactions")

# =========================
# Mood/genre filter
# =========================
MOOD_MAP = {
    "romance":   ["romance", "love", "relationship"],
    "mystery":   ["mystery", "thriller", "detective", "crime"],
    "adventure": ["adventure", "action", "travel", "fantasy"],
    "motivated": ["self-help", "business", "productivity", "inspirational"],
    "sad":       ["drama", "tragedy", "literary"],
    "happy":     ["humor", "comedy", "light"],
    "scifi":     ["science fiction", "sci-fi", "sf"],
    "fantasy":   ["fantasy", "magic", "dragon"],
    "history":   ["history", "historical"],
    "nonfiction":["non-fiction", "biography", "memoir", "essays"]
}

def derive_allowed_categories(mood_text: str):
    mood_text = (mood_text or "").strip().lower()
    if not mood_text:
        return None
    allow = set()
    for key, cats in MOOD_MAP.items():
        if key in mood_text:
            for c in cats: allow.add(c)
    # add tokens from free text
    for tok in mood_text.replace(",", " ").split():
        if len(tok) >= 3:
            allow.add(tok)
    return allow if allow else None

allowed_cats = derive_allowed_categories(MOOD_TEXT)

def build_candidate_mask():
    cand = np.ones(n_items, dtype=bool)
    if allowed_cats:
        cat_series = pd.Series(items_cat, dtype=str).str.lower().fillna("")
        m = np.zeros(n_items, dtype=bool)
        for token in allowed_cats:
            m |= cat_series.str.contains(token, regex=False).values
        if m.sum() >= 50:
            cand &= m
        else:
            print("[WARN] Mood filter too restrictive; using all items.")
    return cand

candidate_mask = build_candidate_mask()

# =========================
# Popularity (fallback/blend)
# =========================
item_pop = np.asarray(Rb.getnnz(axis=0)).astype(np.float32)
if item_pop.max() > 0:
    item_pop /= item_pop.max()

# =========================
# Helpers â€” CSR normalization (FIXED to return CSR)
# =========================
def row_l2norms_csr(M: csr_matrix):
    sq = np.array(M.power(2).sum(axis=1)).ravel()
    norms = np.sqrt(np.maximum(sq, 1e-12))
    return norms

def normalize_rows_csr(M: csr_matrix):
    """Return CSR with each row L2-normalized and norms array."""
    norms = row_l2norms_csr(M)
    inv = 1.0 / norms
    Mn = M.multiply(inv.reshape(-1,1))
    return Mn.tocsr(), norms   # ensure CSR

# Normalize implicit matrix rows â†’ cosine space
Rb_norm, row_norms = normalize_rows_csr(Rb)

# =========================
# Build target user sparse row (normalized)
# =========================
start, end = Rb_norm.indptr[target_u], Rb_norm.indptr[target_u+1]
row_idx = Rb_norm.indices[start:end]
row_dat = Rb_norm.data[start:end]
u_vec = csr_matrix((row_dat, (np.zeros_like(row_idx), row_idx)), shape=(1, n_items))

# =========================
# Cosine similarity to all users (sparse * sparseT)
# =========================
sims = Rb_norm.dot(u_vec.T).toarray().ravel()  # (n_users,)
sims[target_u] = 0.0

# Keep only top-K similar users to speed up
if SIM_TOPK < len(sims):
    top_idx = np.argpartition(-sims, SIM_TOPK)[:SIM_TOPK]
    sims_mask = np.zeros_like(sims, dtype=bool)
    sims_mask[top_idx] = True
    sims_filtered = np.where(sims_mask, sims, 0.0)
else:
    sims_filtered = sims

# =========================
# Score items: sim^T * Rb  (user-based CF)
# =========================
scores_cf = sims_filtered @ Rb    # (items,)
scores_cf = np.array(scores_cf).ravel().astype(np.float32)

# Blend CF and popularity
scores = BLEND_W * scores_cf + (1.0 - BLEND_W) * item_pop

# Exclude already seen + apply candidate mask
seen_start, seen_end = Rb.indptr[target_u], Rb.indptr[target_u+1]
seen_items = set(Rb.indices[seen_start:seen_end].tolist())

scores_masked = scores.copy()
if len(seen_items) > 0:
    scores_masked[list(seen_items)] = -1e9
scores_masked[~candidate_mask] = -1e9

# Guard if fewer candidates than TOP_N
k = min(TOP_N, max(1, int((scores_masked > -1e9).sum())))
top_idx = np.argpartition(-scores_masked, k-1)[:k]
top_idx = top_idx[np.argsort(-scores_masked[top_idx])]

# =========================
# Build results table
# =========================
def trunc(s, n=40):
    s = s or ""
    return s if len(s) <= n else s[:n-1] + "â€¦"

rows = []
for it in top_idx:
    yr = items_year[it]
    if isinstance(yr, (np.integer, int, np.int64, np.int32)): yr = int(yr)
    rows.append({
        "rank": len(rows)+1,
        "isbn": items_isbn[it],
        "title": items_title[it],
        "author": items_author[it],
        "year": yr,
        "category": items_cat[it],
        "language": items_lang[it],
        "score_cf": float(scores_cf[it]),
        "score_pop": float(item_pop[it]),
        "score_blend": float(scores[it]),
    })

rec_df = pd.DataFrame(rows)

# =========================
# Save CSV + JSON + Plot
# =========================
csv_path  = os.path.join(BASE_DIR, f"recommendations_{USER_ID}.csv")
json_path = os.path.join(BASE_DIR, f"recommendations_{USER_ID}.json")
png_path  = os.path.join(BASE_DIR, f"Recommendations_{USER_ID}.png")

rec_df.to_csv(csv_path, index=False, encoding="utf-8")
with open(json_path, "w", encoding="utf-8") as f:
    json.dump({
        "user_id": USER_ID,
        "mood_text": MOOD_TEXT,
        "top_n": int(k),
        "blend_weight_cf": BLEND_W,
        "sim_topk_users": SIM_TOPK,
        "recommendations": rec_df.to_dict(orient="records")
    }, f, ensure_ascii=False, indent=2)

print(f"[OK] Wrote predictions â†’\n  {csv_path}\n  {json_path}")

plt.figure(figsize=(12, 6))
labels = [trunc(t, 36) for t in rec_df["title"].tolist()]
y = rec_df["score_blend"].values
x = np.arange(len(labels))
plt.bar(x, y)
plt.xticks(x, labels, rotation=45, ha="right")
plt.ylabel("Recommendation Score (blend)")
plt.title(f"Top-{k} Recommendations for User {USER_ID}"
          + (f"  |  Mood filter: {MOOD_TEXT}" if MOOD_TEXT.strip() else ""))
plt.tight_layout()
plt.savefig(png_path, dpi=180, bbox_inches="tight")
plt.show()
print(f"[OK] Saved chart â†’ {png_path}")

# Preview top-10 in console
print("\n=== Top Recommendations ===")
for _, r in rec_df.head(10).iterrows():
    yr = f" ({r['year']})" if isinstance(r['year'], (int, np.integer)) and r['year']>0 else ""
    cat = f" â€” {r['category']}" if isinstance(r['category'], str) and r['category'] else ""
    print(f"{int(r['rank']):>2}. {r['title']}{yr}  by {r['author']}{cat}")
```
