```python
# -*- coding: utf-8 -*-
# Book-Crossing â€” Accuracy graph (Hit@K / Precision@K) + Heatmap (user-activity decile Ã— K)
# FIXED: unpacking bug in held-out loop + added safeguards

import os, pickle
from pathlib import Path
import numpy as np
import h5py
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix

# -------------------------
# Paths
# -------------------------
BASE = r"C:\Users\sagni\Downloads\AI Book Recommender with Mood Detection"
PKL  = str(Path(BASE) / "bx_artifacts.pkl")
H5   = str(Path(BASE) / "bx_interactions.h5")
ACC_PNG = str(Path(BASE) / "Accuracy.png")
HM_PNG  = str(Path(BASE) / "Heatmap.png")

# -------------------------
# Load artifacts
# -------------------------
if not Path(PKL).exists() or not Path(H5).exists():
    raise SystemExit("Artifacts not found. Ensure bx_artifacts.pkl and bx_interactions.h5 exist.")

with open(PKL, "rb") as f:
    ART = pickle.load(f)

with h5py.File(H5, "r") as hf:
    shp = tuple(hf["ratings"].attrs["shape"])
    R = csr_matrix((hf["ratings"]["data"][:],
                    hf["ratings"]["indices"][:],
                    hf["ratings"]["indptr"][:]), shape=shp)

    shp_b = tuple(hf["ratings_binary"].attrs["shape"])
    Rb = csr_matrix((hf["ratings_binary"]["data"][:],
                     hf["ratings_binary"]["indices"][:],
                     hf["ratings_binary"]["indptr"][:]), shape=shp_b)

n_users, n_items = R.shape
print(f"[INFO] Loaded CSR: users={n_users:,}, items={n_items:,}, nnz={R.nnz:,}")

# -------------------------
# Train/Test split (implicit positives = rating>=6) â€” hold out 1 positive per user if available
# -------------------------
Rb = Rb.tocsr()
indptr  = Rb.indptr
indices = Rb.indices
data    = Rb.data

# record (u, start, end) of each user row in the CSR
global_pos_indices = [(u, indptr[u], indptr[u+1]) for u in range(n_users)]

rng = np.random.default_rng(42)
train_mask = np.ones(Rb.nnz, dtype=bool)  # mask over all positives (since Rb only stores positives)

test_pos = [[] for _ in range(n_users)]
heldout_count = 0
for u, start, end in global_pos_indices:   # <-- fixed unpacking
    if end <= start:
        continue
    # All entries in this row are positives (Rb), so any can be held out
    cand_len = end - start
    pick_local = int(rng.integers(0, cand_len))
    picked_global = start + pick_local
    train_mask[picked_global] = False
    test_item_col = indices[picked_global]
    test_pos[u] = [int(test_item_col)]
    heldout_count += 1

print(f"[INFO] Users with a held-out positive: {heldout_count:,}")

# Rebuild TRAIN implicit matrix (CSR) after removing held-out entries
train_data    = data[train_mask]
train_indices = indices[train_mask]

# per-user counts before removal
row_counts = np.diff(indptr).astype(np.int64)
# subtract 1 where we held out
train_row_counts = row_counts - (np.array([1 if tp else 0 for tp in test_pos], dtype=np.int64))
train_row_counts = np.clip(train_row_counts, 0, None)

train_indptr = np.zeros(n_users + 1, dtype=np.int64)
np.cumsum(train_row_counts, out=train_indptr[1:])

Rb_train = csr_matrix((train_data, train_indices, train_indptr), shape=Rb.shape)

# Sanity check
if Rb_train.nnz != int(train_mask.sum()):
    raise SystemExit("Rebuild mismatch: train nnz does not match mask sum.")

# -------------------------
# Popularity recommender on TRAIN
# -------------------------
item_pop = np.asarray(Rb_train.getnnz(axis=0)).astype(np.int64)
pop_rank = np.argsort(-item_pop)  # descending

def recommend_for_user(u, K):
    """Top-K unseen items by popularity for user u."""
    s, e = Rb_train.indptr[u], Rb_train.indptr[u+1]
    seen = set(Rb_train.indices[s:e].tolist())
    out = []
    for it in pop_rank:
        if it not in seen:
            out.append(int(it))
            if len(out) == K:
                break
    return out

# -------------------------
# Evaluation: Hit@K and Precision@K over K in {5..50}
# -------------------------
K_list = list(range(5, 55, 5))

# Eligible users: must have a heldout test item and at least 1 train item
eligible_users = [u for u in range(n_users) if test_pos[u] and (Rb_train.indptr[u+1] - Rb_train.indptr[u]) >= 1]

# Cap to keep runtime reasonable on very large sets (adjust as needed)
MAX_USERS_EVAL = 20000
if len(eligible_users) > MAX_USERS_EVAL:
    eligible_users = list(np.random.default_rng(123).choice(eligible_users, size=MAX_USERS_EVAL, replace=False))

print(f"[INFO] Evaluating on users: {len(eligible_users):,}")

# Activity for heatmap bucketing
user_train_nnz = np.asarray(Rb_train.getnnz(axis=1)).astype(int)
activity = np.array([user_train_nnz[u] for u in eligible_users])

hits_at_k, prec_at_k = [], []
for K in K_list:
    hit = 0
    for u in eligible_users:
        t = test_pos[u][0]
        recs = recommend_for_user(u, K)
        if t in recs:
            hit += 1
    hr = hit / len(eligible_users)
    pr = hr / K
    hits_at_k.append(hr)
    prec_at_k.append(pr)
    print(f"[K={K:>2}] Hit@K={hr:.4f}  Precision@K={pr:.4f}")

# -------------------------
# Plot 1: Accuracy curves
# -------------------------
plt.figure(figsize=(7.5,5))
plt.plot(K_list, hits_at_k, marker="o", label="Hit@K")
plt.plot(K_list, prec_at_k, marker="s", label="Precision@K")
plt.title("Popularity Recommender â€” Accuracy vs K")
plt.xlabel("K")
plt.ylabel("Score")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig(ACC_PNG, dpi=180, bbox_inches="tight")
plt.show()
print(f"[OK] Saved accuracy curves â†’ {ACC_PNG}")

# -------------------------
# Plot 2: Heatmap (user-activity decile Ã— K) â€” HitRate
# -------------------------
if len(activity) < 10:
    # single bucket fallback
    buckets = [np.arange(len(activity))]
    dec_labels = ["All"]
else:
    order = np.argsort(activity)
    buckets = np.array_split(order, 10)
    dec_labels = [f"D{idx+1}\n~{int(np.median(activity[idxs]))} intx."
                  for idx, idxs in enumerate(buckets)]

eligible_users_arr = np.array(eligible_users)
H = np.zeros((len(buckets), len(K_list)), dtype=float)

for bi, idxs in enumerate(buckets):
    if len(idxs) == 0:
        continue
    users_subset = eligible_users_arr[idxs]
    for ki, K in enumerate(K_list):
        hit = 0
        for u in users_subset:
            t = test_pos[int(u)][0]
            recs = recommend_for_user(int(u), K)
            if t in recs:
                hit += 1
        H[bi, ki] = hit / len(users_subset)

plt.figure(figsize=(9, 5.2))
plt.imshow(H, aspect="auto")
plt.colorbar(label="HitRate")
plt.xticks(ticks=np.arange(len(K_list)), labels=[str(k) for k in K_list])
plt.yticks(ticks=np.arange(len(dec_labels)), labels=dec_labels)
plt.xlabel("K")
plt.ylabel("User Activity Decile (by train interactions)")
plt.title("HitRate Heatmap â€” Activity Decile Ã— K")
for i in range(H.shape[0]):
    for j in range(H.shape[1]):
        val = H[i, j]
        plt.text(j, i, f"{val:.2f}", ha="center", va="center",
                 fontsize=8, color=("white" if val > 0.5 else "black"))
plt.tight_layout()
plt.savefig(HM_PNG, dpi=200, bbox_inches="tight")
plt.show()
print(f"[OK] Saved heatmap â†’ {HM_PNG}")

print("\nDone.")
```
